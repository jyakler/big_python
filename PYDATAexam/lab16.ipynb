```python
import pandas as pd
import seaborn as sns
import numpy as np
iris = sns.load_dataset("iris")
iris_x = iris.loc[:,['sepal_length', 'sepal_width',
            'petal_length', 'petal_width']]
import random
```

```python
random.seed(1)
for col in range(4):
    iris_x.iloc[[random.sample(range(len(iris)), 20)], col] = float('nan')
iris_x.head(10)
```

```python
df=iris_x.head(10)
df=df.dropna(axis=0)
df
```

```python
df=iris_x.head(11)
df=df.dropna(axis=0,thresh=3)
df
```

```python
df=iris_x.head(15)
df=df.dropna(subset=['sepal_length','sepal_width'],axis=0)
df
```

```python
df=iris_x.head(10)
df=df.fillna(0)
df
```

```python
df=iris_x.head(15)
df=df.fillna(df.max())
df.head(10)
```

```python
df=iris_x.head(10)
df=df.fillna(10)
df.head(10)
```

```python
df=iris_x.head(10)
df=df.fillna(method='ffill')
df.head(10)
```
