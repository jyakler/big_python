```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={
 '1분기':[500,690,1100,1500,1990,1020],
    '2분기':[450,700,1030,1650,2020,1600],
    '3분기':[520,820,1200,1700,2300,2200],
    '4분기':[610,900,1380,1850,2420,2550]
}
df=pd.DataFrame(data,index=[2015,2016,2017,2018,2019,2020])
```

```python
print(df)
```

```python
from matplotlib import rc,font_manager
path='font/MaplestoryBold.ttf'
font_name=font_manager.FontProperties(fname=path).get_name()
rc('font',family=font_name)
plt.rcParams['axes.unicode_minus']=False
```

```python
label=['2015년','2016년','2017년','2018년','2019년','2020년']
plt.figure(figsize=(10,6))
plt.title('시각화 과제1',fontsize=30,color='m')
plt.xticks(df.index,labels=label)
plt.plot(df)
plt.legend(df,loc='upper left')
plt.savefig('output/hw1.png')
```

```python
label=['2015년','2016년','2017년','2018년','2019년','2020년']
plt.figure(figsize=(10,6))
plt.plot(df.T,'^-')
plt.legend(label,loc='upper left')
plt.savefig('output/hw2.png')
```

```python
plt.figure(figsize=(10,6))
plt.pie(df.sum(),autopct='%.1f%%',labels=df.columns,explode=[0.05,0.05,0.05,0.05],shadow=True,colors=['r','g','brown','purple'])

plt.savefig('output/hw3.png')
```

```python
plt.figure(figsize=(10,6))
# plt.figure().patch.set_facecolor('white')
plt.pie(df.T.sum(),autopct='%.1f%%',labels=label,colors=['r','g','brown','purple','yellow','#ff9999'])
# labels=df.T.colums.map(lambda x : str(x)+"년")


plt.savefig('output/hw4.png')
```
