```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
data=pd.read_csv('data/cctv_seoul.csv')
data.set_index('구별',inplace=True)
```

```python
from matplotlib import rc,font_manager
path='font/MaplestoryBold.ttf'
font_name=font_manager.FontProperties(fname=path).get_name()
rc('font',family=font_name)
plt.rcParams['axes.unicode_minus']=False
```

```python
mycolormap = cm.get_cmap("rainbow", 9)
mycolor = [mycolormap(i) for i in range(9)]
plt.figure(figsize=(13,9))
plt.title('시각화 과제(1) 각 구의 CCTV 설치 현황')
plt.ylabel('구이름')
plt.xlabel('각 구에 설치된 CCTV 댓수')
plt.ylim(-0.5,len(data)-.5)
plt.barh(data.index,data['CCTV수'],color=mycolor)

plt.savefig('output/hw5.png')
```

```python
mycolormap = cm.get_cmap("rainbow", 9)
mycolor = [mycolormap(i) for i in range(9)]
plt.figure(figsize=(13,9))
graph=data['CCTV수'].sort_values(ascending=False).plot(kind='bar',title="시각화 과제(2) 각 구의 CCTV 설치 현황",color=mycolor)
graph.set_xlabel('구이름')
graph.set_ylabel('각 구에 설치된 CCTV 댓수')
plt.savefig('output/hw6.png')
```

```python
data['CCTV비율']=data['CCTV수']/data['인구수']*100
data['노인비율']=data['고령자']/data['인구수']*100
```

```python
plt.figure(figsize=(16, 10)) 
mycolormap = cm.get_cmap("Set3", 9)
mycolor = [mycolormap(i) for i in range(9)]
plt.subplot(2,1,1)
plt.title('각 구의 인구수 대비 CCTV비율')
plt.ylabel('CCTV비율')
plt.xlim(-0.5,len(data)-.5)
plt.bar(data.index,data['CCTV비율'],color=mycolor)

plt.subplot(2,1,2)
mycolormap = cm.get_cmap("Set3", 25)
mycolor = [mycolormap(i) for i in range(25)]
plt.title('각 구의 인구수 대비 고령자 비율')
plt.ylabel('고령자 비율')
plt.xlim(-0.5,len(data)-.5)
plt.plot(data['노인비율'],color='gold')
plt.scatter(data.index,data['노인비율'],color=mycolor)
plt.savefig('output/hw7.png')
```
